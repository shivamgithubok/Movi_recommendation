```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
movies = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
credits = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.columns
```

```python
movies.shape
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])

    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
import ast
def convert_cast(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i['name'])
        counter +=1

    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
def convert_crew(text):
    l =[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else [])
```

```python
movies.head(2)
```

```python
def remove_space(word):
    l =[]
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast'] =  movies['cast'].apply(remove_space)
movies['crew'] =  movies['crew'].apply(remove_space)
movies['genres'] =  movies['genres'].apply(remove_space)
movies['keywords'] =  movies['keywords'].apply(remove_space)
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags']  = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df.iloc[0]['tags']
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df['tags'].shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') 
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
def recommend(movie):
    # Check if the movie exists in the dataset
    if movie not in new_df['title'].values:
        print(f"'{movie}' not found in the dataset.")
        return

    # Get the index of the movie
    index = new_df[new_df['title'] == movie].index
    if index.empty:
        print(f"Index for '{movie}' not found.")
        return
    index = index[0]

    # Check similarity matrix dimensions
    if index >= len(similarity):
        print(f"Index {index} is out of bounds for similarity matrix.")
        return

    # Calculate distances and recommend top 5 movies
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    print(f"Top recommendations for '{movie}':")
    for i in distance[1:6]:  # Skip the first (self-match)
        print(new_df.iloc[i[0]]['title'])
```

```python
recommend('Spider-Man')
```
